```python
def crop_image(image_file, filename):
    img = cv2.imread(image_file)
    image_ = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
    image_copy = image_.copy()
    lower = np.array([3,3,3])
    higher = np.array([254,254,254])
    mask=cv2.inRange(image_, lower, higher)
    
    
    contours, hierarchy = cv2.findContours(image= mask,
                                           mode=cv2.RETR_EXTERNAL,
                                           method=cv2.CHAIN_APPROX_NONE)
    sorted_contours=sorted(contours, key=cv2.contourArea, reverse= True) 

    cont_img = cv2.drawContours(image=image_, contours=sorted_contours, contourIdx=0, 
                              color=(0,255,0),thickness=3)
    c = sorted_contours[0]
    x, y, w, h = cv2.boundingRect(c) 
    cv2.rectangle(image_, (x, y), (x + w, y + h), (0, 255, 0), 5)
    cropped_img=image_copy[y:y + h + 1, x:x + w + 1]
    plt.axis('off')
    plt.imshow(cropped_img)
    destination = "./archive/cropped_images/"
    plt.savefig(os.path.join(destination, filename), format='jpg', dpi='figure', bbox_inches='tight', pad_inches=0)
```

```python
train_data = train_data.reset_index()
del train_data['index']
```

```python
for i in range(len(train_data)):
    img = train_data['fundus image'][i]
    crop_image(os.path.join(image_path, img), img)
```

```python
image_path_1 = "./archive/cropped_images"

for i in range(len(train_data)):
    img = train_data['fundus image'][i]
    image_1 = cv2.imread(os.path.join(image_path_1, img))
    resized_image = cv2.resize(image_1, (224, 224))
    dest = "./archive/resized_images"
    cv2.imwrite(os.path.join(dest, img), resized_image)
```

```python
    
image_path_2 = "./archive/resized_images"

for i in range(len(train_data)):
    img = train_data['fundus image'][i]
    image_2 = cv2.imread(os.path.join(image_path_2, img))
    img_blur_ = cv2.GaussianBlur(image_2, (3, 3), 0)
    dest = "./archive/blurred_images"
    cv2.imwrite(os.path.join(dest, img), img_blur_)
```
